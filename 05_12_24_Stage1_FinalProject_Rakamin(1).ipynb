# **DATA PREPROCESSING**

```python
# Menghitung jumlah nilai null di setiap kolom sebelum preprocessing
null_values_before = listings.isnull().sum()

# Menangani Missing Values

# Mengisi nilai kosong pada kolom summary, space, notes, dan neighborhood_overview
for col in ['summary', 'space', 'notes', 'neighborhood_overview']:
    most_frequent_text = listings[col].mode()[0]  # Menghitung teks yang paling sering muncul
    listings[col] = listings[col].fillna(most_frequent_text)  # Imputasi tanpa inplace

# Mengisi nilai kosong pada kolom license
listings['license'] = listings['license'].fillna("Tidak Diperlukan")  # Imputasi untuk license

# Mengisi nilai kosong pada kolom square_feet dengan median
listings['square_feet'] = listings['square_feet'].fillna(listings['square_feet'].median())  # Imputasi dengan median

# Mengisi nilai kosong untuk kolom harga
price_columns = ['price', 'weekly_price', 'monthly_price', 'security_deposit', 'cleaning_fee']
for col in price_columns:
    listings[col] = listings[col].replace({'\$': '', ',': ''}, regex=True).astype(float)  # Menghapus simbol dan mengubah tipe data
    listings[col] = listings[col].fillna(listings[col].median())  # Imputasi dengan median

# Menghitung jumlah nilai null di setiap kolom setelah preprocessing
null_values_after = listings.isnull().sum()

# Membuat DataFrame untuk menampilkan hasil
nvc = pd.DataFrame(null_values_after.sort_values(), columns=['Total Null Values'])
nvc['Percentage'] = (nvc['Total Null Values'] / listings.shape[0]) * 100
nvc["Data Type"] = [listings[col].dtype for col in listings.columns]
nvc.sort_values(by=["Total Null Values", "Percentage"], ascending=False, inplace=True)

# Menampilkan hasil
nvc.style.background_gradient(cmap='Blues')
```
